```python
# Import statements and setup matplotlib for jupyter
import pandas as pd
import numpy as np # for MinMaxScalar
#import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
#%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/PacktPublishing/Machine-Learning-with-R-Third-Edition/master/Chapter07/letterdata.csv'
df = pd.read_csv(url)
df
```

```python
# Split data into testing and training set. Data does not need to be randomly filtered as it was already as mentioned by author.
X = df.drop('letter', axis=1)
y = df['letter']
# Data is split into an 80% training set and 20% testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=800)
```

```python
SVM_Model = SVC(C=5, gamma='auto', kernel='rbf')
SVM_Model.fit(X_train,y_train)
print(f'Accuracy - : {SVM_Model.score(X_test,y_test): .3f}')
```
